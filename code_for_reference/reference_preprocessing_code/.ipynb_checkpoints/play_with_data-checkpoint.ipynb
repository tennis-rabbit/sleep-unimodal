## Step 1: load raw data and preprocess

```python
import pandas as pd
import numpy as np
import os
from wav2sleep.data.edf import load_edf_data
from wav2sleep.data.txt import parse_txt_annotations
from wav2sleep.data.utils import interpolate_index
from wav2sleep.data.xml import parse_xml_annotations
from wav2sleep.settings import *
from wav2sleep.config import *
#from wav2sleep.settings import ABD, CHAT, ECG, INGEST, MROS, PPG, SHHS, THX, WSC
```

```python
'''
Preprocessing notes:
1. select a time window: here we use 10h 
2. select different frequency for different channel
3. resample using interpolate
4. channel-wise normalization
'''

def process_edf(edf: pd.DataFrame):
    """Process dataframe of EDF data."""
    signals = []

    def _process_edf_column(col, target_index):
        """Process signal column of EDF"""
        if col in edf:
            resampled_wav = interpolate_index(edf[col].dropna(), target_index, method="linear", squeeze=False)
            normalized_wav = (resampled_wav - resampled_wav.mean()) / resampled_wav.std()
            print("col:", col, "length:", normalized_wav.shape)
            signals.append(normalized_wav)

    _process_edf_column(ECG, ECG_SIGNAL_INDEX)
    
    return pd.concat(signals, axis=1).astype(np.float32)


def process(edf_fp: str, label_fp: str, output_fp: str, overwrite: bool = False) -> bool:
    """Process night of data."""
    if os.path.exists(output_fp) and not overwrite:
        logger.debug(f'Skipping {edf_fp=}, {output_fp=}, already exists')
        return False
    else:
        os.makedirs(os.path.dirname(output_fp), exist_ok=True)
    # Process labels
    if label_fp.endswith('.xml'):
        try:
            labels = parse_xml_annotations(label_fp)
        except Exception as e:
            logger.error(f'Failed to parse: {label_fp}.')
            logger.error(e)
            return False
    else:
        labels = parse_txt_annotations(fp=label_fp)
        if labels is None:
            logger.error(f'Failed to parse: {label_fp}.')
            return False
    labels = labels.reindex(TARGET_LABEL_INDEX).fillna(-1)
    # Check for N1, N3 or REM presence. (Recordings with just sleep-wake typically use N2 as sole sleep class)
    stage_counts = labels.value_counts()
    
    if stage_counts.get(1.0) is None and stage_counts.get(3.0) is None and stage_counts.get(4.0) is None:
        
        logger.error(f'No N1, N3 or REM in {label_fp}.')
        output_fp = output_fp.replace('.parquet', '.issues.parquet')
    edf = load_edf_data(edf_fp, columns=EDF_COLS, raise_on_missing=False)
    waveform_df = process_edf(edf)
    output_df = pd.concat([waveform_df, labels], axis=1)
    output_df.to_parquet(output_fp)
    return True
```

```python
annotation_path = '/scratch/besp/shared_data/shhs/polysomnography/annotations-events-nsrr/shhs1/shhs1-205804-nsrr.xml'
# edf_path = '/scratch/besp/shared_data/shhs/polysomnography/edfs/shhs1/shhs1-205373.edf'

edf_path = '/scratch/besp/shared_data/shhs/polysomnography/edfs/shhs1/shhs1-205373.edf'
output_path = '/u/ztshuai/ondemand/wav2sleep/test/test.parquet'
EDF_COLS = [ECG]

MAX_LENGTH = 60 * 60 * 10  # Recording length in seconds (trimmed to 10h)
TARGET_LABEL_INDEX = pd.Index(np.arange(0, MAX_LENGTH + 1, 30.0)[1:])
ECG_SIGNAL_INDEX = pd.Index(np.arange(0, MAX_LENGTH + 1e-9, 30 / FREQ_ECG)[1:])  # ~ 34 Hz

process(edf_path, annotation_path, output_path)
```

## Step 4: Statistic Analysis of Preprocessed Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import STL
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from scipy.signal import welch
import warnings


warnings.filterwarnings("ignore")


df = pd.read_parquet('/u/ztshuai/ondemand/wav2sleep/test/test.parquet')


print("=== DataFrame Info ===")
df.info()
print("\n=== Null Values per Column ===")
print(df.isna().sum())


print("\n=== Descriptive Statistics ===")
print(df.describe())


print("\n=== Head of DataFrame ===")
print(df.head())


for col in ['ECG', 'ABD', 'THX']:
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df[col].dropna(), bins=100, kde=True)
    plt.title(f'{col} Distribution')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col].dropna())
    plt.title(f'{col} Boxplot')
    plt.tight_layout()
    plt.show()

```

```python
# === 7. Compute mean & std of ECG by Sleep Stage ===

if 'Stage' in df.columns:
    grouped = df.groupby('Stage')['ECG']
    print("\n=== ECG Mean by Stage ===")
    print(grouped.mean())
    print("\n=== ECG Std by Stage ===")
    print(grouped.std())

    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Stage', y='ECG', data=df.reset_index().dropna(subset=['Stage']))
    plt.title('ECG Distribution by Sleep Stage')
    plt.xlabel('Sleep Stage')
    plt.ylabel('ECG Amplitude')
    plt.show()

# === 8. Rolling statistics (ECG 5-minute window) ===
# Assume the index represents seconds; convert to timedelta so that string-based rolling can be used directly

df_ecg = df[['ECG']].copy()
df_ecg.index = pd.to_timedelta(df_ecg.index, unit='s')
df_ecg_rolling = df_ecg.rolling(window='300s', min_periods=1).mean()

plt.figure(figsize=(12, 4))
plt.plot(df_ecg.index, df_ecg['ECG'], alpha=0.3, label='Original ECG')
plt.plot(df_ecg_rolling.index, df_ecg_rolling['ECG'], color='red', label='5-min Rolling Mean')
plt.title('ECG with 5-minute Rolling Mean')
plt.xlabel('Time')
plt.ylabel('ECG (Z-score)')
plt.legend()
plt.show()

```

```python
# === 9. Trend/seasonal decomposition (replace the previous rolling mean with STL for detailed decomposition) ===
# Here, the period should be set to 1024 points, corresponding to one 30-second epoch with 1024 samples
# The sampling rate is approximately fs = 1024/30 Hz, so the period (i.e., cycle length) can be set to ~1024

stl = STL(df_ecg['ECG'].dropna(), period=1024)
result = stl.fit()
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(result.trend)
plt.title('ECG Trend Component')
plt.subplot(3, 1, 2)
plt.plot(result.seasonal)
plt.title('ECG Seasonal Component')
plt.subplot(3, 1, 3)
plt.plot(result.resid)
plt.title('ECG Residual Component')
plt.tight_layout()
plt.show()

# === 10. Autocorrelation & Partial Autocorrelation (ECG, using the first 20000 points as an example) ===

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plot_acf(df_ecg['ECG'].dropna().iloc[:20000], lags=100, ax=plt.gca(), title='ECG ACF')
plt.subplot(1, 2, 2)
plot_pacf(df_ecg['ECG'].dropna().iloc[:20000], lags=100, ax=plt.gca(), title='ECG PACF')
plt.tight_layout()
plt.show()

# === 11. Spectral analysis (ECG Welch Power Spectral Density) ===

fs = 1024 / 30 
f, Pxx = welch(df_ecg['ECG'].dropna().values, fs=fs, nperseg=2048)
plt.figure(figsize=(8, 4))
plt.semilogy(f, Pxx)
plt.title('ECG Power Spectral Density (Welch)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Density')
plt.show()
```

```python
# === 12. Cross-correlation (ECG vs ABD) ===

df_abd = df[['ABD']].copy()
df_abd.index = pd.to_timedelta(df_abd.index, unit='s')
# 只取两列同时非空的连续时段
common = df_ecg.join(df_abd, how='inner').dropna()
ecg_vals = common['ECG'].values
abd_vals = common['ABD'].values
lags = np.arange(-500, 501)  # ±500 个样本点
corrs = [
    np.corrcoef(
        ecg_vals[max(0, lag):len(ecg_vals)-max(0, -lag)],
        abd_vals[max(0, -lag):len(abd_vals)-max(0, lag)]
    )[0, 1]
    for lag in lags
]

plt.figure(figsize=(8, 4))
plt.plot(lags / fs, corrs)  # 把 lag 转换成秒显示
plt.title('Cross-correlation: ECG vs ABD')
plt.xlabel('Lag (s)')
plt.ylabel('Correlation Coefficient')
plt.grid(True)
plt.show()
```
